```python
from datetime import datetime
class Account:
    def __init__(self, acc_no, name, balance=0, limit=0):
        self.acc_no = acc_no
        self.name = name
        self.balance = balance
        self.limit = limit
        self.history = []
    def check_balance(self):
        return f"Balance: ₹{self.balance}"
    def deposit(self, amount):
        self.balance += amount
        self.history.append(f"Deposit ₹{amount} at {datetime.now()}")
        return f"Deposited ₹{amount}. {self.check_balance()}"
    def withdraw(self, amount):
        if amount > self.limit and self.limit > 0:
            return f"Exceeds limit of ₹{self.limit}."
        if amount <= self.balance + (self.limit if self.limit < 0 else 0):
            self.balance -= amount
            self.history.append(f"Withdraw ₹{amount} at {datetime.now()}")
            return f"Withdrawn ₹{amount}. {self.check_balance()}"
        return "Insufficient funds."
    def show_history(self):
        return "\n".join(self.history) if self.history else "No transactions yet."
def main():
    types = {"savings": (50000, 0), "current": (0, -10000), "checking": (20000, 0)}
    acc_type = input("Account type (savings/current/checking): ").lower()
    if acc_type not in types:
        print("Invalid type!")
        return
    name = input("Name: ")
    acc_no = input("Account number: ")
    balance = float(input("Initial balance: "))
    limit = types[acc_type][1] if acc_type == "current" else types[acc_type][0]
    account = Account(acc_no, name, balance, limit)
    while True:
        choice = input("\n1. Check Balance\n2. Deposit\n3. Withdraw\n4. Show History\n5. Exit\nChoose: ")
        if choice == "1":
            print(account.check_balance())
        elif choice == "2":
            print(account.deposit(float(input("Amount to deposit: "))))
        elif choice == "3":
            print(account.withdraw(float(input("Amount to withdraw: "))))
        elif choice == "4":
            print(account.show_history())
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid choice.")
if __name__ == "__main__":
    main()
```
